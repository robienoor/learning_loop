```python
def getdiff(a,b):
    return np.square(np.subtract(a, b)).mean()
```

```python
training_labels = []

for idx, combo in enumerate(combinations):
    indices = [i for i, x in enumerate(ratings) if x == scores_rounded[idx]]
    print(indices)
    if len(indices) > 1:
        relevant_ranks = ranks[indices]
        diffs = np.array([getdiff(combo, rank) for rank in relevant_ranks])
        selected_diff_idx = np.argmin(diffs)
        selected_rank = indices[selected_diff_idx]
        training_labels.append(ranks[selected_rank])
    else:
        training_labels.append(ranks[indices[0]])
```
